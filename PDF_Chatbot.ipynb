# Pdf loader

```python
loader = PyPDFLoader("environment.pdf")
pages = loader.load()
```

```python
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=256,chunk_overlap=0,separators=[" ",",","\n","."]
)
texts=text_splitter.split_documents(pages)
texts[0]
```

```python
_text=[]
for i in range(len(texts)):
  _text.append(texts[i].page_content)
_text[0]
```

```python
embeddings=HuggingFaceEmbeddings()
```

# Storing document in the chromaDB vector base

```python
db=Chroma.from_documents(texts,embeddings)
```

```python
def similar_docs(query):
    similar_doc=db.similarity_search(query,k=1)
    context=similar_doc[0].page_content
    return context
```

```python
similar_docs('Types of Air pollution')
```

```python
template="""
You are an artificial intelligence assistant. The assistant gives helpful, detailed, and polite answers to the user's questions. Below is some information.
{context}

Based on the above information only, answer the below question.

{question}
"""
def get_answer(query):
  prompt=PromptTemplate.from_template(template)
  llm_chain=LLMChain(prompt=prompt,llm=llm)
  print(llm_chain.run({"context":similar_docs(query),"question": query}))
```

```python
query="what is the Cause of Marine pollution?"
get_answer(query)
```
